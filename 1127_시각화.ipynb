## adl, fall 3축 그래프

```python
df_adl = pd.read_csv('/content/D01_SA01_R01.csv', header = None)
df_fall = pd.read_csv('/content/F01_SA01_R01.csv', header = None)
```

```python
plot_adl_x = df_adl[0].tolist()
plot_adl_y = df_adl[1].tolist()
plot_adl_z = df_adl[2].tolist()

plot_fall_x = df_fall[0].tolist()
plot_fall_y = df_fall[1].tolist()
plot_fall_z = df_fall[2].tolist()
```

```python
plt.figure(figsize=(6,5))
plt.rcParams['figure.figsize'] = (15, 10)      # 사이즈
plt.rcParams['lines.linewidth'] = 0.5          # 선 굵기 

plt.title("ADL의 X, Y, Z축 그래프", pad=15, size=15)
plt.axis([0, 2000, -1500, 1500])                # x_min, x_max . y_min, y_max
# plt.set_xticks([1, 2.5, 3.75, 5, 6.25, 7.5, 8.75, 10])

plt.plot(plot_adl_x,'orange')
plt.plot(plot_adl_y,'darkorange')
plt.plot(plot_adl_z,'peru')

plt.xlabel('단위: 5ms', labelpad=7)
plt.ylabel('센서 값', labelpad=7)
plt.rcParams['xtick.labelsize'] = 30
plt.rcParams['ytick.labelsize'] = 30
plt.tick_params(axis='both', which='major', labelsize=10)
plt.tick_params(axis='both', which='minor', labelsize=10)
plt.legend(['X축', 'Y축', 'Z축'])     


plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.rcParams['figure.figsize'] = (15, 10)      # 사이즈
plt.rcParams['lines.linewidth'] = 0.5          # 선 굵기 

plt.title("Fall의 X, Y, Z축 그래프", pad=15, size=15)
plt.axis([0, 2000, -1500, 1500])                # x_min, x_max . y_min, y_max

plt.plot(plot_fall_x,'orange')
plt.plot(plot_fall_y,'darkorange')
plt.plot(plot_fall_z,'peru')

plt.xlabel('단위: 5ms', labelpad=7,)
plt.ylabel('센서 값', labelpad=7)
plt.rcParams['xtick.labelsize'] = 30
plt.rcParams['ytick.labelsize'] = 30
plt.tick_params(axis='both', which='major', labelsize=10)
plt.tick_params(axis='both', which='minor', labelsize=10)
plt.legend(['X축', 'Y축', 'Z축'])     

plt.show()
```

## 누적 합 그래프

```python
df_sum_adl = pd.read_csv('/content/sum_adl.csv', header = None)
df_sum_fall = pd.read_csv('/content/sum_fall.csv', header = None)
```

```python
adl_cusum = df_sum_adl[1][1000:2000]
fall_cusum = df_sum_fall[1][1000:2000]
# df_cusum.shape
```

```python
plt.figure(figsize=(6,5))
plt.rcParams['figure.figsize'] = (15, 10)      # 사이즈
plt.rcParams['lines.linewidth'] = 1.5          # 선 굵기 

plt.title("ADL, Fall 가속도계 비교", pad=15, size=15)

plt.plot(adl_cusum,'grey')
plt.plot(fall_cusum,'darkorange')

plt.xlabel('단위: 5ms', labelpad=15, size='large')
plt.ylabel('가속도', labelpad=15, size='large')
plt.rcParams['xtick.labelsize'] = 30
plt.rcParams['ytick.labelsize'] = 30
plt.tick_params(axis='both', which='major', labelsize=10)
plt.tick_params(axis='both', which='minor', labelsize=10)
# plt.legend(['ADL', 'Fall'],  bbox_to_anchor=(1, 0.5))     
plt.legend(['ADL', 'Fall'],  loc='center right')     


plt.show()
```
