```python
import pandas as pd
import numpy as np
import csv

f = open('../Dataset/SA01/D01_SA01_R01.csv', 'r')
rdr = csv.reader(f)
file = []
for line in rdr:
    if ';' in line[-1]:
        line[-1] =  line[-1].replace(';','')
    file.append(line)
f.close()

data = pd.DataFrame(file)
data = data[[0, 1, 2]]

col_1_r = data[0].to_numpy().astype(int)
col_2_r = data[1].to_numpy().astype(int)
col_3_r = data[2].to_numpy().astype(int)

col_1 = col_1*col_1
col_2 = col_2*col_2
col_3 = col_3*col_3
new_1 = np.sqrt(col_1 + col_2 + col_3)

newdf_1 = pd.DataFrame(new_1)

import matplotlib.pyplot as plt
plt.figsize=(10000, 5000)
plt.plot(col_1_r[:2400])
plt.plot(col_2_r[:2400])
plt.plot(col_3_r[:2400])
```
