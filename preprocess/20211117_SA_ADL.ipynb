```python
list = []
final_SA01_A = pd.DataFrame(list)

for path in SA01_A_paths:
    data = pd.read_csv(path, header=None)
    data = data.iloc[:, :3]
    rows, cols = data.shape

    sliced_df = data[0 : 160]
    
    col_1_r = sliced_df[0].to_numpy().astype(int)
    col_2_r = sliced_df[1].to_numpy().astype(int)
    col_3_r = sliced_df[2].to_numpy().astype(int)

    col_1 = col_1_r*col_1_r
    col_2 = col_2_r*col_2_r
    col_3 = col_3_r*col_3_r
    
    C_nine = np.sqrt(col_1 + col_2 + col_3)
    list_C_nine = pd.DataFrame(C_nine).transpose().round()
    final_SA01_A = final_SA01_A.append(list_C_nine)

# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)

row, col = final_SA01_A.shape
target_col = row * [1]
final_SA01_A.loc[:,'target'] = target_col

final_SA01_A
```

```python
list = []
final_SA08_F = pd.DataFrame(list)

for path in SA08_F_paths:
    data = pd.read_csv(path, header=None)
    data = data.iloc[:, :3]
    rows, cols = data.shape
    for col in range(cols):
        for row in range(rows):
            if data[col][row] < 0:
                data[col][row] *= -1
    max_value1 = data[1000:][0].max()
    max_value2 = data[1000:][1].max()
    max_value3 = data[1000:][2].max()
    
    max_idx1 = data[ data[0] == max_value1 ].index[0]
    max_idx2 = data[ data[1] == max_value2 ].index[0]
    max_idx3 = data[ data[2] == max_value3 ].index[0]
    min_idx = min(max_idx1, max_idx2, max_idx3)
    sliced_df = data[min_idx - 200 : min_idx - 40]
    
    col_1_r = sliced_df[0].to_numpy().astype(int)
    col_2_r = sliced_df[1].to_numpy().astype(int)
    col_3_r = sliced_df[2].to_numpy().astype(int)

    col_1 = col_1_r*col_1_r
    col_2 = col_2_r*col_2_r
    col_3 = col_3_r*col_3_r
    
    C_nine = np.sqrt(col_1 + col_2 + col_3)
    list_C_nine = pd.DataFrame(C_nine).transpose().round()
    final_SA08_F = final_SA08_F.append(list_C_nine)


# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)

row, col = final_SA08_F.shape
target_col = row * [1]
final_SA08_F.loc[:,'target'] = target_col

final_SA08_F
```

```python
list = []
final_SA17_F = pd.DataFrame(list)

for path in SA17_F_paths:
    data = pd.read_csv(path, header=None)
    data = data.iloc[:, :3]
    rows, cols = data.shape
    for col in range(cols):
        for row in range(rows):
            if data[col][row] < 0:
                data[col][row] *= -1
    max_value1 = data[1000:][0].max()
    max_value2 = data[1000:][1].max()
    max_value3 = data[1000:][2].max()
    
    max_idx1 = data[ data[0] == max_value1 ].index[0]
    max_idx2 = data[ data[1] == max_value2 ].index[0]
    max_idx3 = data[ data[2] == max_value3 ].index[0]
    min_idx = min(max_idx1, max_idx2, max_idx3)
    sliced_df = data[min_idx - 200 : min_idx - 40]
    
    col_1_r = sliced_df[0].to_numpy().astype(int)
    col_2_r = sliced_df[1].to_numpy().astype(int)
    col_3_r = sliced_df[2].to_numpy().astype(int)

    col_1 = col_1_r*col_1_r
    col_2 = col_2_r*col_2_r
    col_3 = col_3_r*col_3_r
    
    C_nine = np.sqrt(col_1 + col_2 + col_3)
    list_C_nine = pd.DataFrame(C_nine).transpose().round()
    final_SA17_F = final_SA17_F.append(list_C_nine)


# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)

row, col = final_SA17_F.shape
target_col = row * [1]
final_SA17_F.loc[:,'target'] = target_col

final_SA17_F
```
